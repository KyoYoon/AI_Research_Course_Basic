```python
import pandas as pd
```

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
grid = GridSearchCV() # 처음에 estimator를 넣어버려야 되서 안 됨 => estimator를 바꾸어 가면서 적용시키고 싶은데 안 됨 
```

```python
grid = GridSearchCV(LogisticRegression(),{'C':[1,2,3,4,5]})
grid.fit(data.data,data.target)
```

```python
grid.cv_results_ # best로 시작하는 값 외에 cv_result 도 포함 
```

```python
pd.DataFrame(grid.cv_results_)
```

```python
# cv_results_ 를 dataFrame 으로 만들고  
# data가 어느 정도 필요한 지 알기 위해 learning curve 와 validation curve를 써서 파악 
```

```python
# 기본 가정 포함하여 해당 값 보여주기 -> cv_result_ => 편차(std_test_score)가 커진다는 것은 데이터가 더 많이 필요하다는 증거  
# 최고의 값만 보여주기 -> best 로 시작하는 변수 
```

```python
# grid search cv 는 알고리즘 1개밖에 안 되서 비교하기 어려움 
# 그래서 이를 편하게 위한 클래스를 만드는 게 좋다. --> BaseEstimator를 상속한 Dummy class 생성 
# 위 방법 외에 search grid(미리 만들어진 라이브러리)를 쓰면 동시에 두 개의 estimator를 비교할 수 있다. 
# pipe line하고 붙여도 estimator를 1개만 써야 하기 때문에 이것도 문제가 있다.
# 최선의 알고리즘과 hyper parameter를 동시에 찾고 싶은데 이를 위해 search grid 를 쓴다. 
# search grid 의 문제점은 업그레이드가 안 된다. 
```

```python
from sklearn.base import BaseEstimator
```

```python
class Dummy(BaseEstimator):
    def fit(self):
        pass
    def score(self):
        pass
```

```python
from sklearn.pipeline import Pipeline, make_pipeline
```

```python
pipe = Pipeline([('clf',Dummy())])
```

```python
param_grid = [{'clf':[LogisticRegression()]}, 
              {'clf':[KNeighborsClassifier()],
               'clf__n_neighbors':[3,4,5,6,7]
              }]
```

```python
grid = GridSearchCV(pipe,param_grid)
```

```python
grid.fit(data.data,data.target)
```

```python
grid.cv_results_
```

```python
pd.DataFrame(grid.cv_results_)
```
