```python
from sklearn.feature_extraction import DictVectorizer # dictionary 형태를 Vectorize하는 것 
```

```python
train = [
    {"user":"1","item":"5","age":19},
    {"user":"2","item":"43","age":33},
    {"user":"3","item":"20","age":55},
    {"user":"4","item":"10","age":20}
]
# key는 mutable한 것이 온다. 
```

```python
v = DictVectorizer() # feature extraction <- 자연어 처리 
```

```python
X = v.fit_transform(train) # fit_transform => feature scaling 포함  
```

```python
X
```

```python
dir(X)
```

```python
dir(v) # inverse_transform을 쓰면 거꾸로 바꿔줌 
```

```python
# python 3.6 부터 뒤에서부터 바꾸어서 나오는 것으로 바뀌었다.
# 자연어 처리는 sparse matrix 구조로 되어있다. 
v.inverse_transform(X) 
```
