```python
data_iris = sns.load_dataset('iris')
```

```python
data_iris
```

```python
knn = KNeighborsClassifier()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_iris.iloc[:,:-1],data_iris.iloc[:,-1])
```

```python
knn.fit(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,knn.predict(X_test)) 
# y_test: () / knn.predict(X_test):  -> numpy => scikit  . 
```

```python
sklearn_evaluation.plot.confusion_matrix(y_test,knn.predict(X_test),normalize=True)
# normalize    1 => 100% / 0.88 => 88%  / 1.0 => 100%
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,knn.predict(X_test)))
```
