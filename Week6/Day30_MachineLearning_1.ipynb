```python
# hyper parameter search 방법 
# grid search cv vs randomized search cv 
# grid search cv -> knn 과 다르게 미리 영역을 다 정해놓고 찾는 방법 
# cross validation -> hyper parameter 와 model algorithm을 찾기 위함이며 overfitting을 막기 위함이다. 
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
```

```python
accuracy = []
for i in range(1,20):
    knn = KNeighborsClassifier(i)
    knn.fit(train_X,train_y)                   #- hold out 방식 
    accuracy.append(knn.score(test_x,test_y))  #-
    #accuracy.append(np.mean(cross_val_score(knn,data.iloc[:,:-1],data.iloc[:,-1],cv=10))) # cross validation
    
# cross validation 은 outlier까지 학습을 시키는 문제점이 있다. 
```

```python
data_iris = sns.load_dataset('iris')
```

```python
data_iris
```

```python
knn = KNeighborsClassifier()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_iris.iloc[:,:-1],data_iris.iloc[:,-1])
```

```python
knn.fit(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,knn.predict(X_test)) 
# y_test: 행(실제) / knn.predict(X_test): 열 -> numpy => scikit에서는 반대로 만들었다. 
```

```python
sklearn_evaluation.plot.confusion_matrix(y_test,knn.predict(X_test),normalize=True)
# normalize 하면 확률 개념처럼 1 => 100% / 0.88 => 88% 맞춤 / 1.0 => 100%
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,knn.predict(X_test)))
```
