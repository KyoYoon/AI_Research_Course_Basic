```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Normalizer
```

```python
X, y = make_blobs()
```

```python
pd_x = pd.DataFrame(X, columns=['x','y'])
pd_y = pd.DataFrame(y, columns=['target'])
```

```python
pd_concat = pd.concat([pd_x,pd_y],axis=1)
```

```python
nn = Normalizer()
```

```python
# 한 개의 열만 넣을 때는 안 되는데 두 개의 열은 된다.  
s = nn.fit_transform(pd_concat[['x','y']]) # dict, property 방식으로 안 됨 -> iloc, columns, fancy indexing
```
