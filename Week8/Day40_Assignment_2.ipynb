```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
iris = sns.load_dataset('iris')
```

```python
def create_model():
    model = Sequential()
    model.add(Dense(8, activation='relu', input_shape=(4,)))
    model.add(Dense(3, activation='softmax'))
    model.compile('adam','categorical_crossentropy',metrics=['accuracy'])
    return model
```

```python
iris
```

```python
iris.species.value_counts()
```

```python
iris_pd = pd.concat([iris,pd.get_dummies(iris.species)],axis=1)
```

```python
iris_pd.drop(columns='species',inplace=True)
```

```python
iris_pd
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris_pd.iloc[:,:-3],iris_pd.iloc[:,-3:])
```

```python
ks = KerasClassifier(create_model,epochs=200,batch_size=32) # epoch을 KerasClassifier에 넣어야 적용된다.
```

```python
ks.fit(X_train,y_train)
```

```python
cross_val_score(ks,X_train,y_train,cv=10)
```
