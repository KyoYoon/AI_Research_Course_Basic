```python
# one hot encoding
data = pd.concat([data,pd.get_dummies(data.species)],axis=1)
```

```python
data.drop(columns='species',inplace=True)
```

```python
data
```

```python
mms = MinMaxScaler()
```

```python
mms.fit_transform(data.iloc[:,:-3])
```

```python
# label encoding 
#X_train, X_test, y_train, y_test = train_test_split(mms.fit_transform(data.iloc[:,:-1]),data.iloc[:,-1])
# multi layer perceptron 은 one hot encoding + min max scaler 적용해야 됨 
X_train, X_test, y_train, y_test = train_test_split(mms.fit_transform(data.iloc[:,:-3]),data.iloc[:,-3:])
```

```python
# perceptron 
```

```python
perceptron = Perceptron()
```

```python
perceptron.fit(X_train,y_train) # one hot encoding으로는 안 되고 label encoding으로는 된다. 
# linear 안에 perceptron이 있다. 
# linear regression 보다 peceptron 이 정확도가 높다. 
# one hot encoding의 경우 node 개수가 달라서 (node 개수 한 개 따지는데) 
# one hot encoding은 안됨 (one hot encoding시 node 개수는 3개)
```

```python
# multi layer perceptron 
multiperc = MLPClassifier()
```

```python
multiperc.fit(X_train,y_train) # label encoding 은 됨 
```

```python
multiperc.fit(X_train,y_train) # one hot encoding + min max scaler 도 됨 
```

```python
multiperc.score(X_test,y_test)
```
