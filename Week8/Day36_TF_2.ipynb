```python
# one hot encoding
iris = pd.concat([iris,pd.get_dummies(iris.species)],axis=1)
```

```python
iris.drop(columns='species',inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris.iloc[:,:-3],iris.iloc[:,-3:])
```

```python
model = Sequential()
```

```python
layer1 = Dense(4,input_shape=(4,))
layer2 = Dense(3) # output 3개
model.add(layer1)
model.add(layer2)
#model.compile('adam','categorical_crossentropy') 
# sparse_categorical_crossentropy
model.compile('adam','categorical_crossentropy') 
#model.fit(iris.)
```

```python
model.fit(X_train,y_train)
```

```python
model.evaluate(X_test,y_test)
```
