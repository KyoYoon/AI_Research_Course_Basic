```python
import numpy as np 
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris # f -> 대용량, m -> 로직 증명, l -> 작은 데이터 load_iris => 함수  
```

```python
iris_pd = pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
iris_pd
```

```python
# 기계학습용 데이터 종류 
# 어떤 데이터가 남은 데이터로 표현이 가능하면 supervised learning을 쓸 수 있다. 
# 예측할 대상을 target 또는 class 라고 한데 이 데이터는 target 이 있다.  
```

```python
target_pd = pd.DataFrame(iris.target, columns=['target'])
```

```python
target_pd
```

```python
# Seaborn 을 이용해서 concat를 쓸 것이다.
```

```python
data_iris = pd.concat([iris_pd,target_pd], axis=1)
```

```python
data_iris
```

```python
data_iris['target'] = data_iris['target'].astype('category')
```

```python
# 위의 것을 보고 어떤 알고리즘을 쓸까 결정 
# classification algorithm을 쓰기로 결정 
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier() # hyper parameter -> 기본 파라메터를 바꾸면 성능이 바뀌는 파라메터 
```

```python
# 통계쪽에서 학습시키는 용어가 fit(A,B)  A: 행(학습시킬 데이터), B: 열(target) 
knn.fit(data_iris.iloc[:,:-1],data_iris.iloc[:,-1])
```

```python
data_iris.info()
```

```python
knn.predict([[3,3,3,3]])
```
