```python
# 전체 크기는 눈에 안 보이는 영역에서 정한다. 즉 plt.figure에서만 정할 수 있다. 
plt.style.use('tableau-colorblind10') # theme 설정 
plt.rcParams['font.family'] = 'KoPubWorldBatang_Pro'
plt.figure(figsize=(7,5), frameon=True) # 눈에 안 보이는 영역 설정 (figsize -> 영역 크기) 
plt.axes() # Artist 객체 상속받아서 그려짐 -> 선언이 안 되도 그래프는 나옴 
plt.title('문근영')
plt.ylabel('문')
plt.xlabel('문')
# plt.yticks([]) # 값이 있으니까 그려짐 
#plt.xticks([]) # 값이 없으니까 안 그려짐 
plt.grid(True)
plt.plot([1,2,3],[4,5,6], label='line1')
plt.plot([1,2,3],[6,5,4], label='line2')
plt.legend() # label 을 쓰고 legend를 쓴다. 
# plot.show는 jupyter에서는 할 필요 없음 
```
