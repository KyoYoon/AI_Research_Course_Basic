```python
import pandas as pd
```

```python
data = pd.read_csv('weather.txt',sep='\t')
```

```python
data.reset_index(inplace=True)
```

```python
data
```

```python
data_long = pd.wide_to_long(data,'d','index','test') 
# melt 또는 wide_to_long 으로 wide format을 long format으로 바꾼다. 
```

```python
data_long
```

```python
def five(df):
    print(df.info())
    print(df.describe())
    print(df.head())
    print(df.tail())
    print(df.sample())
```

```python
five(data_long)
```
