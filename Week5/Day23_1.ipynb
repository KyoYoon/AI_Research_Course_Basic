```python
im = Image.open('gettyimages-1012960656.jpg')
```

```python
im
```

```python
dir(im)
```

```python
im.height, im.width
```

```python
imtoarr = np.array(im) # numpy 포맷으로 변환 
```

```python
imtoarr.shape # 3 -> 3차원 데이터 ( 행: 720 - height, 열: 1280 - width, 차원: 3 - 3가지 채널(R, G, B))
```

```python
imtoarr.flatten() # copy 형식이라 원본은 변경되지 않고 컴퓨터가 이해할 수 있게 1차원 배열로 변환시킴 
```

```python
s = Image.fromarray(imtoarr)
```

```python
s
```

```python
type(s)
```

```python
# 모든 이미지는 배열 형태로 만들어져 있어서 호환하는데 문제가 없음 
```
