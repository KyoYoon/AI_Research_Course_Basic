```python
tips = sns.load_dataset('tips')
```

```python
g = sns.FacetGrid(tips, col='sex',hue='smoker') # ,hue='smoker' -> 담배피는 여부로 색깔 구분 
g.map(plt.scatter,'tip','total_bill')
g.add_legend()
```

```python
g = sns.FacetGrid(tips,row="smoker",col="time",margin_titles=True)
g.map(sns.regplot,"size","total_bill",color=".3")
g.add_legend()
```

```python
ordered_days = tips.day.value_counts().index
g = sns.FacetGrid(tips, row="day", row_order=ordered_days, height=1.7, aspect=4)
g.map(sns.distplot, "total_bill",hist=False, rug=True)
```

```python
# FacetGrid 는 잘라서 보여주는 그래프 
# Matplotlib 로 그린다면 tips.day.unique 와 len(tips.day.unique)
```

```python
fig,ax = plt.subplots(tips.day.nunique(),1)
```

```python
iris = sns.load_dataset("iris")
g = sns.PairGrid(iris) # Grid 가 column 명으로 쪼개짐 
g.map(plt.scatter) # 안에 들어갈 그래프를 매핑 -> 그래프를 그려줄 형태만 넣어주면 됨 
```

```python
# 내가 원하는 row와 column으로 잘라주고 자른 것에 어떤 그래프를 그릴지 지정해주는 게 FacetGrid
# column 기준으로 전체를 잘라주는 것이 PairGrid 
```

```python
h = sns.PairGrid(iris,hue="species") # PairPlot
# h.map(plt.scatter)
h.map_diag(plt.hist) # 좌상단부터 우하단까지 대각선에 있는 공간은 히스토그램으로 그려라 
h.map_offdiag(plt.scatter) # 대각선 바깥 공간은 scatter plot으로 그려라. 
h.add_legend()
```
