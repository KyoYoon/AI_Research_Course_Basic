```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
tips.info()
```

```python
tips['size'] = tips['size'].astype('category') # numpy 에 size 가 존재
```

```python
tips.info()
```

```python
tips.rename({'size':'size_'}, axis=1, inplace=True) # 열 개념 
```

```python
tips
```

```python
# 날짜별 팁 평균 
tips.groupby('day').mean()[['tip']] # [[]] -> data frame / [] -> series 
```

```python
# crosstab 
pd.crosstab(tips.day,tips.smoker,margins=True)
```

```python
pd.crosstab(tips.day,tips.smoker,margins=True,values=tips.tip,aggfunc=np.mean)
```

```python
t = tips.groupby(['day','smoker']).mean()[['tip']]
```

```python
t
```

```python
# 행이 x축 / index가 x축
```

```python
t.plot.bar(stacked=True)
```

```python
t.stack() # column은 index로 만들어줌
```

```python
t.unstack() # index를 column으로 만들어줌 
```

```python
t.unstack().plot.bar()
```

```python
t.unstack().plot.bar(stacked=True)
```
